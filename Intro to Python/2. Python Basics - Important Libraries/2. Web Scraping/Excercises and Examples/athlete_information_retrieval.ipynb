# NHL Players

```python
url_nhl ="https://www.quanthockey.com/nhl/players/"
```

```python
import string

a_to_z_lowercase = string.ascii_lowercase
list1 = []
for i in a_to_z_lowercase:
    list1.append(f"{url_nhl}{i}.php")
```

```python
combine_dfs={}
for num, l in enumerate(a_to_z_lowercase[:]):
    try:
        df=pd.read_html(list1[num])[0]
        combine_dfs[l]=df.drop(columns=['Unnamed: 0'])
    except:
        continue
```

```python
combine_dfs
```

```python
result_df = pd.concat(combine_dfs)
```

```python
result_df #returns all players EVER to have played in the NHL
```
